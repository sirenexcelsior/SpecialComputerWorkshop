```python
# 定义多项式环
R = PolynomialRing(QQ, 'x, y')
x, y = R.gens()

# 定义方程系统
equations = [
    x**2 + y**2 - 1,
    x - y**2
]

# 构造Groebner基
I = ideal(equations)
G = I.groebner_basis()

# 检查Groebner基的性质
if G == [1]:
    print("该方程系统有无穷多个解。")
else:
    print("该方程系统仅有有限个解。")
```

```python
def Groebner_is(equations):
    # 定义多项式环
    R = PolynomialRing(QQ, 'x, y')
    x, y = R.gens()
    # 构造Groebner基
    I = ideal(equations)
    G = I.groebner_basis()
    return G == [1]
```

```python
# 定义方程系统
equations = [
    x^2 + y^2 - 1,
    x - y^2
]

Groebner_is(equations)
```
